## Global stats

```python
sql('''
SELECT max(total_cases) AS total_cases, max(total_vaccinations) AS total_vaccinations,
    max(hosp_patients) AS hosp_patients, max(total_deaths) AS total_deaths
FROM covid_vaccinations AS v
JOIN covid_deaths AS d
    ON v.date = d.date AND v.location = d.location
''').df()
```

## Let's look at the following questions:
* How many were infected out of total population?
* How many were hospitalized?
* How many died out of total cases?

```python
sql('''
SELECT v.date, v.location, population, total_cases, total_vaccinations, hosp_patients, total_deaths,
    round(total_cases/population*100, 2) AS case_percentage,
    round(total_deaths/total_cases*100, 2) AS death_percentage
FROM covid_vaccinations AS v
JOIN covid_deaths AS d
    ON v.date = d.date AND v.location = d.location
WHERE v.continent IS NOT NULL
ORDER BY 1, 2 DESC
''').df()
```

## Which countries were most infected with highest mortality?

```python
sql('''
SELECT location, population, max(total_cases) AS total_cases, max(total_deaths) AS total_deaths,
    round(max(total_cases)/population*100, 2) AS case_percentage,
    round(max(total_deaths) / max(total_cases) * 100, 2) AS death_percentage
FROM covid_deaths
WHERE continent IS NOT NULL
GROUP BY location, population
ORDER BY 4 DESC
''').df()
```

## Which continents have highest mortality?

```python
sql('''
SELECT location, population, max(total_deaths) AS total_deaths
FROM covid_deaths
WHERE continent IS NULL AND location NOT IN ('World', 'European Union', 'International')
GROUP BY location, population
ORDER BY 3 DESC
''').df()
```

## Time-series death counts

```python
sql('''
SELECT date, sum(new_cases) AS new_cases, sum(new_deaths) AS new_deaths,
    round(sum(new_deaths) / sum(new_cases) * 100, 2) AS death_percentage
FROM covid_deaths
WHERE continent IS NOT NULL AND new_cases IS NOT NULL
GROUP BY date
ORDER BY 1
''').df()
```

## Cumulative total vaccination count by countries

```python
sql('''
WITH a AS (
    SELECT v.location, v.date, population, new_vaccinations,
        sum(new_vaccinations) OVER (PARTITION BY v.location ORDER BY v.location, v.date) AS total_vaccinations
    FROM covid_vaccinations AS v
    JOIN covid_deaths AS d
        ON v.date = d.date AND v.location = d.location
    WHERE v.continent IS NOT NULL
    ORDER BY 1, 2)
SELECT *, round(total_vaccinations/population*100, 2) AS vaccination_percentage
FROM a
''').df()
```

## Stringency index in different countries

```python
sql('''
PIVOT (
    SELECT date, location, stringency_index
    FROM covid_vaccinations
    WHERE continent IS NOT NULL
    ORDER BY 1, 2)
ON location
USING avg(stringency_index)
''').df()
```

## Hospital capacity vs patients
A strain metric is calculated using the formula below:
$$
\text{strain\_percentage} = \left( \frac{\text{max(hosp\_patients)}}{\left(\frac{\text{population}}{1000} \times \text{hospital\_beds\_per\_thousand}\right)} \right) \times 100
$$

```python
sql('''
SELECT v.location, population, hospital_beds_per_thousand, max(hosp_patients) hosp_patients,
    max(hosp_patients) / (population / 1000 * hospital_beds_per_thousand) * 100 AS 'hospital_strain_percentage'
FROM covid_vaccinations AS v
JOIN covid_deaths AS d
    ON v.date = d.date AND v.location = d.location
WHERE v.continent IS NOT NULL
GROUP BY v.location, population, hospital_beds_per_thousand
ORDER BY 5 DESC
''').df()
```
